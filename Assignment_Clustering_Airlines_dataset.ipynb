# K-Mean Clustering

```python
airline
```

```python
airline3=airline.drop(['ID#'],axis=1)
airline3
```

```python
# Normalize heterogenous numerical data by using Standard Scaler
airline2_std=StandardScaler().fit_transform(airline3)
```

```python
# Use Elbow Graph to find optimum number of  clusters (K value) from K values range
# The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS 
# random state can be anything from 0 to 42, but the same number to be used everytime,so that the results don't change.
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(airline2_std)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# Build Cluster algorithm using K=4

```python
# Cluster algorithm using K=4
clusters4=KMeans(4,random_state=30).fit(airline2_std)
clusters4
```

```python
clusters4.labels_
```

```python
# Assign clusters to the data set
airline4=airline3.copy()
airline4["clusters4id"]=clusters4.labels_
airline4
```

```python
clusters4.cluster_centers_
```

```python
airline4.groupby('clusters4id').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(airline4['clusters4id'],airline4['Balance'], c=clusters4.labels_) 
```

# DBSCAN

```python
data=airline.drop(['ID#'],axis=1)
data
```

```python
data_norm=StandardScaler().fit_transform(data)
data_norm
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=22)
dbscan.fit(data_norm)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

-1 are the noise points in our dataset and the rest are the clusters

```python
# Adding clusters to dataset
data['clusters']=dbscan.labels_
data
```

```python
data.groupby('clusters').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(data['clusters'],data['Balance'], c=dbscan.labels_) 
```
